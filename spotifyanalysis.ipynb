```python
df.replace("", np.nan, inplace=True)
df.replace(" ", np.nan, inplace=True)
df = df.dropna()
```

```python
df.isna().sum()
```

```python
plt.scatter(df['track_popularity'], df['artist_popularity'])
plt.xlabel("track_popularity")
plt.ylabel("artist_popularity")
plt.title("track_popularity vs artist_popularity")
plt.show()
```

**Average track popularity**

```python
df['track_popularity'].mean()
```

**Most popular track**

```python
df.loc[df['track_popularity'].idxmax()][['track_name','artist_name','track_popularity']]
```

**Least popular track**

```python
df.loc[df['track_popularity'].idxmin()][['track_name','artist_name','track_popularity']]
```

**Longest duration track**

```python
df.loc[df['track_duration_min'].idxmax()][['track_name','artist_name','track_duration_min']]
```

**Track popularity vs followers**

```python
plt.scatter(df['artist_followers'], df['track_popularity'])
plt.xlabel("Artist Followers")
plt.ylabel("Track Popularity")
plt.show()
```

**Popularity distribution**

```python
plt.hist(df['track_popularity'], bins=30)
plt.xlabel("Popularity")
plt.show()
```

 **Release year vs popularity trend**

```python
df.groupby('album_release_date')['track_popularity'].mean().plot(figsize=(12,4))
plt.show()
```

**Explicit vs non-explicit genres**

```python
df.groupby(['explicit','artist_genres'])['track_name'].count()
```

**Artist with highest average track popularity**

```python
df.groupby('artist_name')['track_popularity'].mean().sort_values(ascending=False).head(10)
```
